```python
df = pd.read_csv('input/train_ver2.csv',usecols=train_cols)
```

```python
df_june = pd.read_csv('input/test_ver2.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_may = df[df['fecha_dato']=='2016-05-28']

df.drop('fecha_dato', axis=1, inplace=True)
```

```python
dfm = pd.merge(df_june,df_may, how='left', on=['ncodpers'], suffixes=('', '_prev'))

prevcols = [col for col in dfm.columns if '_ult1_prev' in col]
#currcols = [col for col in dfm.columns if '_ult1' in col and '_ult1_prev' not in col]

for col in prevcols:
    dfm[col].fillna(0, inplace=True)
```

```python
#dfm = dfm[dfm[currcols].sum(axis=1) >0]
```

```python
prevcols2 = [col for col in dfm.columns if '_prev' in col and col not in prevcols]
dfm.drop(prevcols2, axis=1, inplace=True)
```

```python
dfm.rename(columns=lambda x: x.replace('_ult1', '_ult1_prev'), inplace=True)
```
